# **Project Summary -**

Mental health has become a major concern in today’s workplace, especially in the technology sector, where high pressure, long hours, and fast-paced changes are common. Even with growing awareness, mental health is still a sensitive subject, often linked to stigma and little support from organizations. This project aims to conduct an Exploratory Data Analysis (EDA) on the Mental Health in Tech Survey (2014) dataset to better understand employees’ mental health conditions, workplace attitudes, and the support systems available.

The dataset includes 1,259 survey responses and 27 variables, covering demographic data, mental health history, treatment-seeking behavior, workplace policies, and employee views on mental health. Since the dataset is survey-based and mostly categorical, the analysis focuses on cleaning the data, exploring categories, and uncovering relationships instead of predictive modeling.

These steps ensured that the dataset was clean, consistent, and ready for analysis.

Univariate analysis revealed that most respondents are in the early to mid-career age group, with many reporting a history of mental health issues or seeking treatment. The analysis also indicated that many have a family history of mental illness, which affects their treatment-seeking behavior. The distribution of company size showed that many respondents work in small to mid-sized organizations, where formal mental health policies are often lacking.

Bivariate and multivariate visualizations offered deeper insights into workplace attitudes. One key finding was that employees with a family history of mental illness are more likely to seek treatment, while age alone does not strongly influence treatment behavior. Several visualizations indicated that mental health often disrupts work productivity, highlighting the business importance of mental well-being. The analysis also found that many organizations do not provide mental health benefits or that employees are unaware of them, pointing to a communication and policy gap.

The correlation heatmap further underscored that factors such as the availability of mental health benefits, anonymity protection, and wellness programs are more closely linked to openness and treatment-seeking than demographic factors. This indicates that company policies and workplace culture significantly affect employee mental health outcomes. Additionally, insights from interview-related questions showed that stigma remains, as most candidates feel uncomfortable discussing mental health in job interviews.

Overall, the project illustrates that mental health challenges are common in the tech industry, but supportive workplace environments can greatly reduce stigma and enhance employee engagement. The findings support the notion that investing in mental health initiatives is beneficial for employees and leads to positive business outcomes, such as higher productivity, lower turnover, and a stronger organizational culture.

In conclusion, this EDA emphasizes the need for data-driven decision-making in tackling workplace mental health. By providing comprehensive mental health benefits, ensuring confidentiality, training managers, and encouraging open communication, organizations can create healthier and more sustainable work environments. This project reinforces that mental health should be prioritized alongside physical health for long-term organizational success.

## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
````

````python
df_wrangle = df.copy()
````

````python
df_wrangle['self_employed'] = df_wrangle['self_employed'].fillna('No')
df_wrangle['work_interfere'] = df_wrangle['work_interfere'].fillna('Not Applicable')
df_wrangle['state'] = df_wrangle['state'].fillna('Not Applicable')
df_wrangle.drop(columns=['comments'], errors='ignore')
# Keeping age within a realistic working range
df_wrangle = df_wrangle[(df_wrangle['Age'] >= 18) & (df_wrangle['Age'] <= 65)]
````

````python
def clean_gender(gender):
    gender = gender.lower()
    if gender in ['male', 'm', 'man', 'cis male', 'male-ish', 'maile']:
        return 'Male'
    elif gender in ['female', 'f', 'woman', 'cis female', 'female-ish']:
        return 'Female'
    else:
        return 'Other'

df_wrangle['Gender'] = df_wrangle['Gender'].apply(clean_gender)
````

````python
df_wrangle = df_wrangle.replace({
    'Yes': 'Yes',
    'No': 'No',
    'Maybe': 'Maybe',
    "Don't know": 'Unknown'
})
````

````python
bins = [18, 25, 35, 45, 55, 65]
labels = ['18-25', '26-35', '36-45', '46-55', '56-65']

df_wrangle['Age_Group'] = pd.cut(df_wrangle['Age'], bins=bins, labels=labels)
````

````python
df_wrangle.info()
````

### What all manipulations have you done and insights you found?

**Manipulations Performed**

Handled missing values by:

* Filling logical defaults (Not Applicable, No)

* Dropping high-null, low-impact column (comments)

Removed unrealistic age values to maintain data quality.

Standardized the Gender column to reduce noise caused by inconsistent entries.

Normalized categorical responses for consistency.

Created an Age_Group feature to simplify age-based analysis.

Ensured the dataset is clean, consistent, and analysis-ready.

**Insights Gained**

Missing values were not random; they depended on respondent context (e.g., country or diagnosis).

Gender had significant inconsistency, which could distort analysis if not standardized.

Most respondents fall within early to mid-career age groups.

Cleaning significantly improves reliability of trends and visualizations.

After wrangling, the dataset becomes suitable for univariate, bivariate, and multivariate analysis.

#### Chart - 14 - Correlation Heatmap

````python
# Correlation Heatmap visualization code
import seaborn as sns
import matplotlib.pyplot as plt

# Create a copy
corr_df = df_wrangle.copy()

# Drop non-numeric / derived categorical columns
corr_df = corr_df.drop(columns=['Age_Group'], errors='ignore')

# Encode categorical columns safely
for col in corr_df.columns:
    if corr_df[col].dtype == 'object':
        corr_df[col] = corr_df[col].astype('category').cat.codes

# Compute correlation matrix
corr_matrix = corr_df.corr()

# Plot heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Mental Health Survey Data")
plt.show()


````

##### 1. Why did you pick the specific chart?

A correlation heatmap is selected to quantify and visualize the strength of relationships between multiple variables at once. It provides a compact summary of how workplace policies, mental health attitudes, and treatment-related factors are interrelated, making it easier to detect patterns that may not be obvious from individual charts.

##### 2. What is/are the insight(s) found from the chart?

The heatmap reveals that workplace support-related variables (such as benefits, care options, anonymity, and wellness programs) show meaningful correlations with treatment-seeking behavior and openness in discussing mental health. It also indicates that the relationship between demographic factors and treatment is weaker compared to organizational and policy-driven factors.

#### Chart - 15 - Pair Plot

````python
# Select relevant columns for pair plot
pairplot_df = df_wrangle[['Age', 'family_history', 'treatment']].copy()

# Encode categorical variables
pairplot_df['family_history'] = pairplot_df['family_history'].map({'Yes': 1, 'No': 0})
pairplot_df['treatment'] = pairplot_df['treatment'].map({'Yes': 1, 'No': 0})

# Pair Plot
sns.pairplot(pairplot_df, hue='treatment')
plt.show()
````

##### 1. Why did you pick the specific chart?

A pair plot is chosen because it allows simultaneous visualization of relationships between multiple variables. It helps in identifying patterns, trends, and separations between groups by showing pairwise relationships and distributions in a single view. In this analysis, it is particularly useful to study how age and family history relate to mental health treatment.

##### 2. What is/are the insight(s) found from the chart?

The pair plot shows that treatment-seeking behavior is more strongly associated with family history of mental illness than with age. Respondents with a family history are clearly more likely to have sought treatment, while age does not show a strong separation between those who sought treatment and those who did not.
